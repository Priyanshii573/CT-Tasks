```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Display the first few rows of the dataset
print("First few rows of the Titanic dataset:\n", df.head())

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot the count of survivors and non-survivors
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Survived', palette='viridis')
plt.title('Count of Survivors and Non-Survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Plot the count of passengers in each class
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass', palette='viridis')
plt.title('Count of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# Plot the distribution of ages of the passengers
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True, palette='viridis')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot the survival rate based on gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sex', hue='Survived', palette='viridis')
plt.title('Survival Rate Based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks([0, 1], ['Female', 'Male'])
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
